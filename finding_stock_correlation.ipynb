```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dataset = pd.read_csv('data/all_stock_close.csv')
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset.set_index('Date', drop=True, inplace=True)
dataset.head(3)
```

```python
prevday_returns = np.log(dataset.shift(1)/dataset.shift(2))
today_returns = np.log(dataset/dataset.shift(1))
```

```python
today_returns.head(5)
```

```python
# Forming dataframe for today_returns
# Deleting the NaN rows
df1 = pd.concat([today_returns], axis=1)
df1 = df1.drop(df1.index[[0,1,-1]], axis=0)
df1.head(5)
```

```python
# Forming dataframe for prevday_returns
# Deleting the NaN rows
df2 = pd.concat([prevday_returns], axis=1)
df2 = df2.drop(df2.index[[0,1,-1]], axis=0)
df2.head(5)
```

```python
# Finding correlation between today_returns of all stocks & the prevday_returns of catalyzing stocks (catstocks)
# Sorting correlation results in descending order
catstocks = ['0001.HK', '0002.HK']
for catstock in catstocks:
    sorted_corr = df1.corrwith(df2[catstock], axis=0).sort_values(ascending=False)
    print(sorted_corr.head(10))
```

```python
# Finding contemporaneous correlation between all stocks & top 5 rising stocks
# Sorting correlation results in descending order
catstocks = ['2008.HK', '0112.HK', '1302.HK', '1771.HK', '1499.HK']
for catstock in catstocks:
    sorted_corr = df1.corrwith(df1[catstock], axis=0).sort_values(ascending=False)
    print(sorted_corr.head(10))
```
