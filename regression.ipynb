```python
import pandas as pd
import numpy as np
import datetime as dt
import statsmodels.api as sm
```

```python
dataset = pd.read_csv('data/all_stock_close.csv')
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset.set_index('Date', drop=True, inplace=True)
dataset.head(3)
```

```python
prevday_returns = np.log(dataset.shift(1)/dataset.shift(2))
today_returns = np.log(dataset/dataset.shift(1))
```

```python
# Forming dataframe for today_returns
# Deleting the NaN rows
df1 = pd.concat([today_returns], axis=1)
df1 = df1.drop(df1.index[[0,1,2,-1]], axis=0)
df1.dropna(inplace=True)
```

```python
# Forming dataframe for prevday_returns
# Deleting the NaN rows
df2 = pd.concat([prevday_returns], axis=1)
df2 = df2.drop(df2.index[[0,1,2,-1]], axis=0)
df2.dropna(inplace=True)
```

```python
df2.head(5)
```

```python
catstocks = ['2008.HK', '0001.HK']

# Creating an empty list for regression results
results = []

# Regressing stocks in df1 against each catstock
for catstock in catstocks:
    X = df2[[catstock]]
    X = sm.add_constant(X)
    print('X has {} np.nan'.format(X.isnull().sum()))
    print(X[X.isnull()])
    for column in df1.columns:
        y = df1[[column]]
        print('y has {} np.nan'.format(y.isnull().sum()))
        model = sm.OLS(y, X).fit()
        results.append(model.params)
        results.append(model.tvalues)
        results.append(model.rsquared)

print(results)
```
